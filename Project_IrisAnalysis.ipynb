```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#reading csv file
df = pd.read_csv("C:/Users/vinodini/OneDrive/Documents/Iris.csv")

#print first 5 rows
df.head()
```

```python
#data visualization

sns.countplot(x='Species' , data=df)
plt.show()
```

```python
#hue--it shows the dependency of the passed data values are to be plotted

sns.scatterplot(x="SepalLengthCm" , y="SepalWidthCm",hue='Species',data=df)
plt.legend(bbox_to_anchor=(1,1),loc=2) 
plt.show()
```

```python
#pairplot--use to see every possible pairing of shape and color at once

sns.pairplot(df.drop(['Id'], axis=1),hue='Species', height=2)
```

```python
#histogram-- allows seeing the distribution of data for various columns

fig, axes=plt.subplots(2,2, figsize=(8,8))
axes[0,0].set_title("Sepel Length")
axes[0,0].hist(df['SepalLengthCm'],bins=7)

axes[0,1].set_title("Sepel Width")
axes[0,1].hist(df['SepalWidthCm'],bins=5)

axes[1,0].set_title("Petal Length")
axes[1,0].hist(df['PetalLengthCm'],bins=6)

axes[1,1].set_title("Petal Width")
axes[1,1].hist(df['PetalWidthCm'],bins=6)
```
